### 1 - Aplicando o algoritmo de busca binária do zero apenas a partir da definição, sem auxílio de IA.

```python
## Tentando aplicar o algoritmo de busca binária apenas a partir da definição do zero sem IA ou auxílios externos:

lista = list(range(1, 101))

target = int(input("Insira um valor entre 1 e 100 a ser buscado na lista: "))

def busca_binaria(lista, alvo):
    global contadorBinario
    contadorBinario += 1
    if lista[len(lista)//2] > alvo:
        novaLista = lista[:len(lista)//2]
    elif lista[len(lista)//2] < alvo:
        novaLista = lista[len(lista)//2 + 1:]
    else:
        return "Encontrei pela busca binária."
    
    return busca_binaria(novaLista, alvo)

def busca_simples(lista, alvo):
    global contadorSimples
    
    for i in lista:
        contadorSimples += 1
        if i == alvo:
            return "Encontrei pela busca simples"

contadorBinario = 0
contadorSimples = 0

busca_binaria(lista, target)
busca_simples(lista, target)
print(f"O algoritmo de busca binária foi executado {contadorBinario} vezes.")
print(f"O algoritmo de pesquisa simples foi executado {contadorSimples} vezes.")
```

- Pontos a melhorar: 
    - Uso de global pode ser evitado.
    - O ideal é utilizar os índices início e fim para conseguir ser guiar na lista, envés de fazer o fatiamento dela, pois as fatias ficam salvas na memória.
    - Lidar com lista vazia.
    - Lidar com valores fora do range do input.

```python
## Aplicando as melhorias acima

lista = list(range(1, 101))

alvo = int(input("Digite um valor entre 1 e 100 para ser procurado na lista: "))

i = 0 
f = len(lista) - 1 # 99
contador = 0

# Deve ser i <= f para que seja feita a última iteração, que seria para testar i == f, ou seja, quando o alvo é encontrado, além disso, se o valor não for encontrado na lista, por exemplo, 101, o i será maior que o f, pois no final o f irá assumir o valor de -1.
while i <= f:
    if lista[(f + i) // 2] > alvo:
        f = ((f + i) // 2) - 1
        contador += 1 
    elif lista[(f + i) // 2] < alvo:
        i = ((f + i) // 2) + 1
        contador += 1
    elif lista[(f + i) // 2] == alvo:
        i = f = (f + i) // 2
        contador += 1
        break

if i > f:
    print("Alvo não encontrado na lista.")
else:
    print(f"Encontrei o valor no índice {i} da lista executando a busca binária {contador} vezes.")
```
