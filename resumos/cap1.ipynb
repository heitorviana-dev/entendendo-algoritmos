## O que é um algoritmo?

Para começarmos a estudar sobre algoritmos precisamos primeiro saber definir o que é um algoritmo. Um algoritmo nada mais é que um conjunto de instruções que realiza uma tarefa, então, por exemplo, a linha de código `print("Hello World!")` é um algoritmo.

## A busca binária:

Vamos começar com a seguinte situação, imagine que você queira procurar o número do seu amigo em uma lista telefônica, a lista de nomes está organizada em ordem alfabética. A primeira solução seria procurar um por um, então se o nome do seu amigo começasse com A seria encontrado rapidamente, se começasse com Z deveria percorrer toda a lista até encontrar o nome do seu amigo.

Concorda comigo que se o nome do seu amigo for o último da lista telefônica você deverá percorrer nome a nome até encontrá-lo? Chamamos esse tipo de algoritmo de busca linear, o qual itera(percorre) a lista e para quando encontra o nome, desse modo se a lista tem 100 nomes a quantidade máxima de operações que podem ser realizadas para encontrar um nome é 100(o último).

Porém, e se envés de percorrermos a lista inteira nós a partissemos em duas e verificassemos se o nome que buscamos está na primeira ou segunda metade, você acha que seria mais eficiente? Seria assim, imagina que estamos procurando o nome Heitor na lista telefônica, primeiro pegamos o valor central e verificamos, o nome central está a frente no alfabeto ou atrás da letra H? Se estiver a frente nós descartamos a segunda metade e continuamos realizando a divisão, se estiver atrás nós descartamos a primeira metade, e se for igual nós paramos ali mesmo e retornamos o número de telefone.

No exemplo abaixo estamos procurando o valor 20 em uma lista ordenada de 7 valores:



Veja que em uma lista de 7 valores nós faríamos no máximo 3 operações, que seriam, dividir, dividir e verificar se é o valor que procuramos. Se tivessemos usado a busca linear teríamos realizado 7 operações.

O nome do algoritmo que acabamos de demonstrar é **Busca Binária**, um algoritmo que funciona para procurar valores em uma lista *ordenada*, lembrando a lista deve ser ordenada.

## Implementação da busca binária:

Agora vamos implementar a busca binária na linguagem python para uma lista de números que vão de 1 a 100, antes de ver a implementação recomendo que tente visualizar esse problema em listas menores como de 1 a 8 e escrever no papel, você pode tentar até a implementar antes de ver a solução e comparar depois.

```python
lista = list(range(1, 101)) # [1, 2, ..., 100]

alvo = int(input("Digite um valor entre 1 e 100: ")) # 100

contador = 0 # Indicar a quantidade de iterações do algoritmo
inicio = 0 # Valor do índice de onde a lista começa
fim = len(lista) - 1 # Valor do índice de onde a lista termina

while inicio <= fim:
    meio = (fim + inicio) // 2 # Índice do valor central
    
    if lista[meio] > alvo:
        fim = meio - 1 # -1 pois o valor central não é o alvo
        contador += 1
    elif lista[meio] < alvo:
        inicio = meio + 1 # +1 pois o valor central não é o alvo
        contador += 1
    elif lista[meio] == alvo:
        inicio = fim = meio
        contador += 1
        break # Finaliza o loop

if inicio > fim: # Caso o alvo não seja encontrado na lista, o fim irá se tornar -1(faça o teste no papel)
    print("O alvo não foi encontrado na lista.")
else:
    print(f"O alvo foi encontrado na posição {inicio} da lista após {contador} iterações utilizando o algoritmo da busca binária.")
```

Para fazermos a implementação utilizamos os índices *início* e *fim* para que possamos percorrer pela lista sem precisar alterá-la (você poderia pensar em usar fatiamento de listas, foi o que pensei na primeira vez, porém fatiar listas utiliza um algoritmo com mesma eficiência da busca linear, logo não faz sentido usá-la) utilizando ambos os índices para indicar onde está o meio da lista.

## Como medir a eficiência de um algoritmo?

Certo, aprendemos sobre a busca binária e sobre a busca linear, você pode estar se perguntando qual o melhor? Para responder isso podemos fazer a seguinte pergunta, como podemos medir o quão eficiente um algoritmo é? Você pode pensar em utilizar o tempo em segundos ou talvez na simplicidade do algoritmo (geralmente algoritmos mais eficientes são mais complicados de se implementar) ou no número de operações. Para respondermos essa pergunta vamos apresentar uma nomenclatura muito utilizada na ciência da computação no âmbito de algoritmos, a **Notação Big O**. 

## Notação Big O

A notação Big O é uma notação que nós informa o quão rápido um algoritmo cresce, e é exatamente assim que medimos a eficiência de um algoritmo.

Imagine a seguinte situação, você está programando um avião, antes de decolar você precisa encontrar um valor em uma lista ordenada para que assim possa fazê-lo sem problemas, você tem 10 segundos, cada operação custo em torno de 1ms, você pensa 1ms é muito pouco vou aplicar a busca linear que é mais fácil, quão esperto você é, a lista, nos testes, possuia em torno de 10 valores. O tempo de operação será de 10ms sendo dentro do limite aceitável, se aplicassemos a busca binária que é mais complicada teríamos um tempo de apenas 4ms, você olha os dois tempos e fala a diferença não é tão grande, menos de três vezes, vou deixar a busca linear pra facilitar pra mim. Chega na hora da decolagem, porém agora a lista contém 1.000.000.000 de valores, tendo agora um tempo de operação de 1.000.000.000ms ou 1.000.000s infringindo, e muito, o limite estabelecido.

Se aplicassemos a busca binária nessa lista, quanto você acha que daria? Para facilitar os seus calcúlos o número de operações é dado por log n na base 2, estou supondo que você sabe o que são logaritmos, se não sabe trate de estudar, porém uma dica, eles nada mais são que o inverso da potenciação (e não eles não são raiz por favor). Utilizando a busca binária obteríamos um tempo de 30ms, a diferença passou de cerca de 3 vezes para 33 milhões de vezes.

Você conseguiu entender que não basta vermos o tempo de operação atual, mas também precisamos analisar o quanto cada algoritmo cresce assim que você aumentar a lista?

O algoritmo de busca linear possui o tempo de operação O(n) enquanto o algoritmo de busca binária possui um tempo de operação O(log n), o que isso representa? Representa que o número de operações que serão realizadas no pior caso de cada algoritmo, ou seja, praticamente uma garantia, te informando qual o máximo número de operações.

Portanto, a notação big O é utilizada para informar o quão rápido um algoritmo cresce e usá-se a notação O() para isso.

## Exemplos de Notações Big O:

As notações O(n) e O(log n) discutidas anteriormente são chamadas de **tempo linear** e **tempo logarítmico** respectivamente. Além desses temos alguns outros tempos muito importantes que seria bom você ir se familiarizando:

    - O(1): Tempo constante

    - O(log n): Tempo logarítmico

    - O(n): Tempo linear

    - O(n*log n): Tempo linearítmico ou apenas nlog n

    - O(nˆ2): Tempo quadrático

    - O(nˆ3): Tempo cúbico

    - O(nˆk): Tempo polinomial

    - O(n!): Tempo fatorial

## Problema do caixeiro viajante

Imagine que você vai viajar para 5 cidades e você quer saber a rota que possui a menor distância, você concorda que para isso nós deveríamos permutar por todas as cidades, calcular a distância de cada rota e no fim pegar a menor? Permutar é simples apenas fazemos 5! que dará 120, ou seja, deveriam ser feita 120 operações para obtermos a melhor rota que passe pelas 5 cidades.

Porém agora, imagine que você queira passar por 6 cidades, veja que o número de operações agora será de 6! ou 720, se fosse 10 cidades teríamos 3.628.800 operações. Veja que o número de operações cresce de maneira exponencial, de modo que se passassemos por 100 cidades nós veríamos o sol colapsar antes de termos a melhor rota. 

Esse é o problema do caixeiro viajante, um problema considerado sem solução, em que podemos apenas obter uma solução aproximada, porém não temos uma solução ótima. O tempo de execução em notação Big O é O(n!).

A partir de agora chamarei o tempo de execução Big O de complexidade Big O, fique atento.
